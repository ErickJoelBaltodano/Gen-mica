# **Alineamiento para el virus del sarampión**
**NOTA:** Antes de correr los diferentes códigos se debe de carga el archivo:

*   sequences.fasta

Al ejecutar este código nos pedirá elegir el archivo para cargar la secuencia del archivo sequences.fasta.

En total son 14 secuencias identificadas por su accession number y el número total de bases que tiene cada secuencia.

```python
!pip install biopython --quiet

import matplotlib.pyplot as plt
import numpy as np
from collections import Counter
from Bio import pairwise2
from google.colab import files
from Bio import SeqIO
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
import pandas as pd
# ==========================
# 1. CARGA DEL FASTA
# ==========================
# Subir archivo
uploaded = files.upload()
filename = next(iter(uploaded))

# Cargar secuencias en memoria
seqs = list(SeqIO.parse(filename, "fasta"))

print("Secuencias cargadas:")
for s in seqs:
    print(s.id, len(s.seq), "nt")

# ============================================================
# 2. TABLA DE LONGITUDES (TODAS LAS SECUENCIAS)
# ============================================================
df_lengths = pd.DataFrame({
    "ID": [s.id for s in seqs],
    "Longitud (nt)": [len(s.seq) for s in seqs]
})
print("\n=== Longitudes de cada secuencia ===")
display(df_lengths)

# ============================================================
# 3. COMPOSICIÓN NUCLEOTÍDICA DE TODAS
# ============================================================
compositions = []
for s in seqs:
    seq_str = str(s.seq).upper()
    counts = Counter(seq_str)
    total = len(seq_str)
    comp = {
        "ID": s.id,
        "A": counts.get("A", 0),
        "T": counts.get("T", 0),
        "G": counts.get("G", 0),
        "C": counts.get("C", 0),
        "%A": counts.get("A", 0)/total*100,
        "%T": counts.get("T", 0)/total*100,
        "%G": counts.get("G", 0)/total*100,
        "%C": counts.get("C", 0)/total*100,
    }
    compositions.append(comp)

df_comp = pd.DataFrame(compositions)
print("\n=== Composición nucleotídica ===")
display(df_comp)

# ============================================================
# 4. ALINEAMIENTO PAREADO ENTRE TODAS LAS SECUENCIAS
# ============================================================
match = 2
mismatch = -1
gap_open = -5
gap_extend = -0.5

resultados = []
n = len(seqs)

print("\n=== Generando alineamientos pareados ===\n")

for i in range(n):
    for j in range(i+1, n):
        s1 = str(seqs[i].seq)
        s2 = str(seqs[j].seq)

        aln = pairwise2.align.globalms(
            s1, s2,
            match, mismatch, gap_open, gap_extend,
            one_alignment_only=True
        )[0]

        a1, a2, score, start, end = aln

        # % identidad
        matches = sum((a == b and a != "-") for a, b in zip(a1, a2))
        overlap = sum((a != "-" and b != "-") for a, b in zip(a1, a2))
        pct_id = matches / overlap * 100

        resultados.append({
            "Seq1": seqs[i].id,
            "Seq2": seqs[j].id,
            "Score": score,
            "%Identidad": pct_id,
            "Longitud_aln": len(a1)
        })

df_align = pd.DataFrame(resultados)
print("\n=== Resultados de alineamientos pareados ===")
# display(df_align)
# Mostrar solo 5 primeras y 5 últimas filas
if len(df_align) > 10:
    display(pd.concat([df_align.head(5), df_align.tail(5)]))
else:
    display(df_align)


# ============================================================
# 5. MATRIZ DE IDENTIDAD ENTRE TODAS LAS SECUENCIAS
# ============================================================

matriz = pd.DataFrame(
    np.zeros((n, n)),
    index=[s.id for s in seqs],
    columns=[s.id for s in seqs]
)

for r in resultados:
    matriz.loc[r["Seq1"], r["Seq2"]] = r["%Identidad"]
    matriz.loc[r["Seq2"], r["Seq1"]] = r["%Identidad"]

print("\n=== MATRIZ DE IDENTIDAD (%) ===")
display(matriz)

plt.figure(figsize=(10,7))
plt.imshow(matriz, cmap="viridis")
plt.colorbar(label="Identidad (%)")
plt.xticks(range(n), matriz.columns, rotation=90)
plt.yticks(range(n), matriz.index)
plt.title("Matriz de identidad entre secuencias")
plt.show()

# ============================================================
# 7. DIFERENCIAS SIGNIFICATIVAS (>= 3 bases) + FILTRO COMPLEMENTO
# ============================================================

print("\n=== DIFERENCIAS SIGNIFICATIVAS ENTRE CEPAS (>= 3 bases) ===\n")

# Mostrar la tabla completa
pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', None)

# Mapa de años
years = {
    "AF411001.1": 2001,
    "KJ659872.1": 2011, "KJ659873.1": 2011, "KJ659874.1": 2011,
    "KP769784.1": 2015, "KP769785.1": 2015,
    "MH042758.1": 2018,
    "MK628192.1": 2019, "MN544303.1": 2019, "MN544304.1": 2019,
    "MN544305.1": 2019, "MN544306.1": 2019, "MN544307.1": 2019,
    "MN544308.1": 2019
}

def detectar_regiones_diferentes(a1, a2):
    diferencias = []
    inicio = None
    for i, (x, y) in enumerate(zip(a1, a2)):
        if x != y:
            if inicio is None:
                inicio = i
        else:
            if inicio is not None:
                diferencias.append((inicio, i - 1))
                inicio = None
    if inicio is not None:
        diferencias.append((inicio, len(a1) - 1))
    return diferencias

# Función para evaluar complementos
comple = str.maketrans("ATGC", "TACG")

def es_complementario(f1, f2):
    return f1.translate(comple) == f2

tabla_dif = []

for i in range(n):
    for j in range(i + 1, n):

        id1 = seqs[i].id
        id2 = seqs[j].id
        year1 = years.get(id1, "NA")
        year2 = years.get(id2, "NA")

        s1 = str(seqs[i].seq)
        s2 = str(seqs[j].seq)

        aln = pairwise2.align.globalms(
            s1, s2,
            match, mismatch, gap_open, gap_extend,
            one_alignment_only=True
        )[0]

        a1, a2, score, start, end = aln

        regiones = detectar_regiones_diferentes(a1, a2)

        for (ini, fin) in regiones:
            longi = fin - ini + 1
            if longi < 3:
                continue

            frag1 = a1[ini:fin + 1]
            frag2 = a2[ini:fin + 1]

            # Omitir diferencias no útiles
            if frag1.startswith("-") or frag2.startswith("-") or frag1.startswith("G-") or frag2.startswith("G-"):
                continue

            # Omitir secuencias que son complementarias
            if es_complementario(frag1, frag2):
                continue

            tabla_dif.append({
                "Seq1": id1,
                "Año1": year1,
                "Seq2": id2,
                "Año2": year2,
                "Inicio": ini,
                "Fin": fin,
                "Longitud": longi,
                "Fragmento_Seq1": frag1,
                "Fragmento_Seq2": frag2
            })

df_dif = pd.DataFrame(tabla_dif)

print("Total de diferencias significativas encontradas:", len(df_dif))
# display(df_dif)
# Mostrar solo 5 primeras y 5 últimas filas
if len(df_dif) > 10:
    display(pd.concat([df_dif.head(5), df_dif.tail(5)]))
else:
    display(df_dif)


# ============================================================
# 8. COMPARACIÓN POR AÑO
# ============================================================

print("\n=== COMPARACIÓN DE DIFERENCIAS POR AÑO ===\n")

# Creamos nueva tabla solo con los años
df_years = df_dif.copy()
df_years["Par_Años"] = df_years["Año1"].astype(str) + " → " + df_years["Año2"].astype(str)

# ------------------------------------------------------------
# 1. Conteo total de mutaciones entre cada par de años
# ------------------------------------------------------------
mut_por_año = df_years.groupby("Par_Años").size().reset_index(name="Mutaciones")

print("\n>>> Total de mutaciones reales entre años:")
display(mut_por_año)

# ------------------------------------------------------------
# 2. Matriz Año vs Año (heatmap base)
# ------------------------------------------------------------

# Una matriz de conteo año1 x año2
matriz_anios = df_years.pivot_table(
    index="Año1",
    columns="Año2",
    values="Longitud",
    aggfunc="count",
    fill_value=0
)

print("\n>>> Matriz Año1 vs Año2 (conteo de mutaciones):")
display(matriz_anios)

plt.figure(figsize=(7,5))
plt.imshow(matriz_anios, cmap="viridis")
plt.title("Mapa de calor de mutaciones por año")
plt.xlabel("Año 2")
plt.ylabel("Año 1")
plt.colorbar(label="Número de mutaciones")
plt.xticks(range(len(matriz_anios.columns)), matriz_anios.columns)
plt.yticks(range(len(matriz_anios.index)), matriz_anios.index)
plt.show()

# ------------------------------------------------------------
# 3. Promedio de extensión de mutaciones por año
# ------------------------------------------------------------
prom_long = df_years.groupby("Año1")["Longitud"].mean().reset_index(name="Promedio_Longitud")

print("\n>>> Promedio de longitud de mutaciones por año (Año1):")
display(prom_long)
```
